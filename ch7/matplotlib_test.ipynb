```python
# Í∞ÑÎã®Ìïú ÏÇ¨Ïö© ÏòàÏãú: Í∫æÏùÄÏÑ† Í∑∏ÎûòÌîÑ Í∑∏Î¶¨Í∏∞
# Í∞ÄÏû• Í∏∞Î≥∏Ï†ÅÏù∏ Í∫æÏùÄÏÑ† Í∑∏ÎûòÌîÑ(Line Plot)Î•º Í∑∏Î¶¨Îäî Í≥ºÏ†ïÏùÄ Îã§ÏùåÍ≥º Í∞ôÏäµÎãàÎã§.
#
# 1. Îç∞Ïù¥ÌÑ∞ Ï§ÄÎπÑ: xÏ∂ïÍ≥º yÏ∂ïÏóê Ìï¥ÎãπÌïòÎäî Îç∞Ïù¥ÌÑ∞Î•º Ï§ÄÎπÑÌï©ÎãàÎã§.
# (Ï£ºÎ°ú Î¶¨Ïä§Ìä∏ÎÇò NumPy Î∞∞Ïó¥ ÏÇ¨Ïö©)
# 2. ÌîåÎ°Ø ÏÉùÏÑ±: plt.plot() Ìï®ÏàòÎ•º ÏÇ¨Ïö©Ìï¥ Îç∞Ïù¥ÌÑ∞Î•º Í∑∏ÎûòÌîÑÎ°ú Í∑∏Î¶ΩÎãàÎã§.
# 3. Í∑∏ÎûòÌîÑ Íæ∏ÎØ∏Í∏∞ (ÏÑ†ÌÉù ÏÇ¨Ìï≠):
# plt.title(), plt.xlabel(), plt.ylabel() Îì±ÏúºÎ°ú Ï†úÎ™©Í≥º Ï∂ï Ïù¥Î¶ÑÏùÑ Ï∂îÍ∞ÄÌï©ÎãàÎã§.
# 4. Í∑∏ÎûòÌîÑ Î≥¥Ïó¨Ï£ºÍ∏∞: plt.show() Ìï®ÏàòÎ•º Ìò∏Ï∂úÌïòÏó¨ ÌôîÎ©¥Ïóê Í∑∏ÎûòÌîÑÎ•º ÌëúÏãúÌï©ÎãàÎã§.

import matplotlib.pyplot as plt

# 1. Îç∞Ïù¥ÌÑ∞ Ï§ÄÎπÑ (xÏ∂ï: Ïõî, yÏ∂ï: ÏõîÎ≥Ñ Îß§Ï∂ú)
months = [1, 2, 3, 4, 5, 6]
sales = [150, 220, 280, 250, 310, 350]

# 2. Í∫æÏùÄÏÑ† Í∑∏ÎûòÌîÑ ÏÉùÏÑ±
plt.plot(months, sales)

# 3. Í∑∏ÎûòÌîÑ Íæ∏ÎØ∏Í∏∞
plt.title("Monthly Sales Trend") # Ï†úÎ™© Ï∂îÍ∞Ä
plt.xlabel("Month")              # xÏ∂ï Ïù¥Î¶Ñ Ï∂îÍ∞Ä
plt.ylabel("Sales (in millions)") # yÏ∂ï Ïù¥Î¶Ñ Ï∂îÍ∞Ä

# 4. Í∑∏ÎûòÌîÑÎ•º ÌôîÎ©¥Ïóê ÌëúÏãú
plt.show()
```

```python
# ÏòàÏãú 5
# ‚ë£ ÏÇ∞Ï†êÎèÑ (Scatter Plot)
# Îëê Î≥ÄÏàò Í∞ÑÏùò Í¥ÄÍ≥ÑÎ•º ÏãúÍ∞ÅÌôîÌï† Îïå ÏÇ¨Ïö©Îê©ÎãàÎã§.
#
# ÏÇ∞Ï†êÎèÑ(Scatter Plot)Îäî
# Îëê Í∞úÏùò ÏÑúÎ°ú Îã§Î•∏ Ïà´Ïûê Î≥ÄÏàò ÏÇ¨Ïù¥Ïùò Í¥ÄÍ≥ÑÎ•º
# ÏãúÍ∞ÅÏ†ÅÏúºÎ°ú Î≥¥Ïó¨Ï£ºÎäî Í∑∏ÎûòÌîÑÏûÖÎãàÎã§.
# üìä Í∞Å Îç∞Ïù¥ÌÑ∞ Ìè¨Ïù∏Ìä∏Î•º xÏ∂ï Í∞íÍ≥º yÏ∂ï Í∞íÏóê Ìï¥ÎãπÌïòÎäî ÏúÑÏπòÏóê Ï†êÏúºÎ°ú Ï∞çÏñ¥ ÎÇòÌÉÄÎÉÖÎãàÎã§.
#
# Ï£ºÎ°ú Îëê Î≥ÄÏàò Í∞ÑÏóê Ïñ¥Îñ§ Ïó∞Í¥ÄÏÑ±Ïù¥ ÏûàÎäîÏßÄ,
# Ï¶â ÏÉÅÍ¥ÄÍ¥ÄÍ≥ÑÎ•º ÌååÏïÖÌïòÎäî Îç∞ ÏÇ¨Ïö©Îê©ÎãàÎã§.
# ÏòàÎ•º Îì§Ïñ¥, 'Í≥µÎ∂Ä ÏãúÍ∞Ñ'Í≥º 'ÏãúÌóò Ï†êÏàò'
# ÏÇ¨Ïù¥Ïùò Í¥ÄÍ≥ÑÎ•º ÏïåÏïÑÎ≥¥Í≥† Ïã∂ÏùÑ Îïå Îß§Ïö∞ Ïú†Ïö©Ìï©ÎãàÎã§.
#
# ## ÏÇ∞Ï†êÎèÑÏùò ÌïµÏã¨ Íµ¨ÏÑ± ÏöîÏÜå
# xÏ∂ï Î≥ÄÏàò: Í¥ÄÍ≥ÑÎ•º ÌååÏïÖÌïòÍ≥†Ïûê ÌïòÎäî Îëê Î≥ÄÏàò Ï§ë ÌïòÎÇòÏûÖÎãàÎã§. (ÎèÖÎ¶Ω Î≥ÄÏàò)
#
# yÏ∂ï Î≥ÄÏàò: Îã§Î•∏ ÌïòÎÇòÏùò Î≥ÄÏàòÏûÖÎãàÎã§. (Ï¢ÖÏÜç Î≥ÄÏàò)
#
# Ï†ê (Point): Í∞ÅÍ∞ÅÏùò Îç∞Ïù¥ÌÑ∞ ÌïòÎÇòÎ•º ÏùòÎØ∏Ìï©ÎãàÎã§.
# Ìïú Ï†êÏùÄ Ìïú ÏåçÏùò (x, y) Í∞íÏùÑ Í∞ÄÏßëÎãàÎã§. (Ïòà: Ïñ¥Îñ§ ÌïôÏÉùÏùò Í≥µÎ∂Ä ÏãúÍ∞ÑÍ≥º ÏãúÌóò Ï†êÏàò)
#
# ## ÏÇ∞Ï†êÎèÑÎ•º ÌÜµÌï¥ Ïïå Ïàò ÏûàÎäî Í≤É
# ÏÇ∞Ï†êÎèÑÏóê Ï∞çÌûå Ï†êÎì§Ïùò Ï†ÑÏ≤¥Ï†ÅÏù∏ Ìå®ÌÑ¥ÏùÑ ÌÜµÌï¥ Îëê Î≥ÄÏàò Í∞ÑÏùò Í¥ÄÍ≥ÑÎ•º ÏßÅÍ¥ÄÏ†ÅÏúºÎ°ú ÌååÏïÖÌï† Ïàò ÏûàÏäµÎãàÎã§.
#
# ÏñëÏùò ÏÉÅÍ¥ÄÍ¥ÄÍ≥Ñ (Positive Correlation)
# Ìïú Î≥ÄÏàòÍ∞Ä Ï¶ùÍ∞ÄÌï† Îïå Îã§Î•∏ Î≥ÄÏàòÎèÑ Ìï®Íªò Ï¶ùÍ∞ÄÌïòÎäî Í≤ΩÌñ•ÏùÑ Î≥¥ÏûÖÎãàÎã§.
# Ï†êÎì§Ïù¥ Ïò§Î•∏Ï™Ω ÏúÑÎ°ú Ìñ•ÌïòÎäî Ìå®ÌÑ¥ÏùÑ Î≥¥ÏûÖÎãàÎã§.
# (Ïòà: Í≥µÎ∂Ä ÏãúÍ∞ÑÏù¥ ÎäòÏñ¥ÎÇ†ÏàòÎ°ù ÏãúÌóò Ï†êÏàòÎèÑ ÎÜíÏïÑÏßÑÎã§.)
#
# ÏùåÏùò ÏÉÅÍ¥ÄÍ¥ÄÍ≥Ñ (Negative Correlation)
# Ìïú Î≥ÄÏàòÍ∞Ä Ï¶ùÍ∞ÄÌï† Îïå Îã§Î•∏ Î≥ÄÏàòÎäî Í∞êÏÜåÌïòÎäî Í≤ΩÌñ•ÏùÑ Î≥¥ÏûÖÎãàÎã§.
# Ï†êÎì§Ïù¥ Ïò§Î•∏Ï™Ω ÏïÑÎûòÎ°ú Ìñ•ÌïòÎäî Ìå®ÌÑ¥ÏùÑ Î≥¥ÏûÖÎãàÎã§.
# (Ïòà: Í≤åÏûÑ ÏãúÍ∞ÑÏù¥ ÎäòÏñ¥ÎÇ†ÏàòÎ°ù ÏãúÌóò Ï†êÏàòÎäî ÎÇÆÏïÑÏßÑÎã§.)
#
# ÏÉÅÍ¥ÄÍ¥ÄÍ≥Ñ ÏóÜÏùå (No Correlation)
# Îëê Î≥ÄÏàò ÏÇ¨Ïù¥Ïóê ÌäπÎ≥ÑÌïú Í¥ÄÍ≥ÑÍ∞Ä Î≥¥Ïù¥ÏßÄ ÏïäÏäµÎãàÎã§. Ï†êÎì§Ïù¥ ÎöúÎ†∑Ìïú Ìå®ÌÑ¥ ÏóÜÏù¥ Ìù©Ïñ¥Ï†∏ ÏûàÏäµÎãàÎã§.
# (Ïòà: ÌÇ§ÏôÄ ÏãúÌóò Ï†êÏàòÎäî Î≥ÑÎã§Î•∏ Í¥ÄÍ≥ÑÍ∞Ä ÏóÜÎã§.)

# 4. ÏÇ∞Ï†êÎèÑ
x = np.random.rand(50)
y = np.random.rand(50)
plt.figure(figsize=(8, 5))

plt.scatter(x, y, color="red", alpha=0.7)

plt.xlabel("X Í∞í", fontproperties=font_prop)
plt.ylabel("Y Í∞í", fontproperties=font_prop)
plt.title("ÏÇ∞Ï†êÎèÑ ÏòàÏ†ú", fontproperties=font_prop)
plt.grid(True)
plt.show()

# Ï£ºÏöî ÏòµÏÖò
#
# alpha=0.7 ‚Üí Ìà¨Î™ÖÎèÑ Ï°∞Ï†à
# color="red" ‚Üí Ï†ê ÏÉâÏÉÅ ÏßÄÏ†ï
#
# üìå Ï∂îÍ∞Ä ÏòµÏÖò
#
# s=100 ‚Üí Ï†ê ÌÅ¨Í∏∞ Ï°∞Ï†à
# c=np.random.rand(50) ‚Üí Ï†ê ÏÉâÏÉÅÏùÑ Í∞íÏóê Îî∞Îùº Îã§Î•¥Í≤å ÏÑ§Ï†ï
# cmap='viridis' ‚Üí Ïª¨Îü¨Îßµ ÏÑ§Ï†ï ('coolwarm', 'plasma', 'cividis')
# edgecolors='black' ‚Üí Ï†ê ÌÖåÎëêÎ¶¨ ÏÉâÏÉÅ ÏßÄÏ†ï
# linewidths=0.5 ‚Üí Ï†ê ÌÖåÎëêÎ¶¨ ÎëêÍªò Ï°∞Ï†à
# alpha=0.6 ‚Üí Ï†ê Ìà¨Î™ÖÎèÑ Ï°∞Ï†à
# plt.colorbar() ‚Üí Ïª¨Îü¨ Î∞î Ï∂îÍ∞Ä (Í∞íÏù¥ ÏÉâÏúºÎ°ú ÌëúÌòÑÎê† Í≤ΩÏö∞)
# plt.xscale('log') ‚Üí XÏ∂ï Î°úÍ∑∏ Ïä§ÏºÄÏùº Ï†ÅÏö©
# plt.scatter(x, y, marker='^') ‚Üí ÎßàÏª§ Ïä§ÌÉÄÏùº Î≥ÄÍ≤Ω ('o', '^', 's', 'd')
# plt.axhline(y=0.5, color='r', linestyle='--') ‚Üí ÌäπÏ†ï yÍ∞í Í∏∞Ï§ÄÏÑ† Ï∂îÍ∞Ä
```

```python
# ÏòàÏãú 7
# ‚ë• ÌûàÌä∏Îßµ (Heatmap)

# Îç∞Ïù¥ÌÑ∞ ÌñâÎ†¨ÏùÑ ÏãúÍ∞ÅÌôîÌï† Îïå ÏÇ¨Ïö©Îê©ÎãàÎã§.
# ÌûàÌä∏Îßµ(Heatmap)ÏùÄ
# Îç∞Ïù¥ÌÑ∞ Í∞íÏùò ÌÅ¨Í∏∞Î•º ÏÉâÏÉÅÏúºÎ°ú ÌëúÌòÑÌïòÎäî ÏãúÍ∞ÅÌôî Í∏∞Î≤ïÏûÖÎãàÎã§.
# üìú Ï£ºÎ°ú ÌëúÎÇò ÌñâÎ†¨ ÌòïÌÉúÏùò Îç∞Ïù¥ÌÑ∞ÏóêÏÑú Í∞íÏù¥ ÎÜíÏùÄ Î∂ÄÎ∂ÑÍ≥º
# ÎÇÆÏùÄ Î∂ÄÎ∂ÑÏùÑ ÌïúÎààÏóê ÌååÏïÖÌï† Ïàò ÏûàÎèÑÎ°ù ÎèÑÏôÄÏ§çÎãàÎã§.
#
# "Ïó¥(Heat) ÏßÄÎèÑ"ÎùºÎäî Ïù¥Î¶ÑÏ≤òÎüº,
# Í∞íÏù¥ ÎÜíÏùÑÏàòÎ°ù Îú®Í±∞Ïö¥ ÏÉâ(Ïòà: Î∂âÏùÄÏÉâ)ÏúºÎ°ú,
# Í∞íÏù¥ ÎÇÆÏùÑÏàòÎ°ù Ï∞®Í∞ÄÏö¥ ÏÉâ(Ïòà: Ìë∏Î•∏ÏÉâ)ÏúºÎ°ú ÎÇòÌÉÄÎÇ¥Îäî Í≤ΩÏö∞Í∞Ä ÎßéÏäµÎãàÎã§.
#
# ## ÌûàÌä∏ÎßµÏùò ÌïµÏã¨ Íµ¨ÏÑ± ÏöîÏÜå
# Í≤©Ïûê (Grid): Îç∞Ïù¥ÌÑ∞Í∞Ä Îã¥Í≤® ÏûàÎäî 2Ï∞®Ïõê Í≥µÍ∞ÑÏúºÎ°ú,
# Í∞Å ÏÖÄ(Ïπ∏)Ïù¥ ÌïòÎÇòÏùò Îç∞Ïù¥ÌÑ∞ Í∞íÏùÑ Í∞ÄÏßëÎãàÎã§. (Í∞ÄÎ°úÏ∂ïÍ≥º ÏÑ∏Î°úÏ∂ïÏùÄ Î≥ÄÏàòÎ•º ÎÇòÌÉÄÎÉÖÎãàÎã§.)
#
# ÏÉâÏÉÅ Ï≤ôÎèÑ (Color Scale):
# Îç∞Ïù¥ÌÑ∞ Í∞íÏùò ÌÅ¨Í∏∞Î•º ÎÇòÌÉÄÎÇ¥Îäî ÏÉâÏÉÅÏùò Î≤îÎ°ÄÏûÖÎãàÎã§.
# Ïù¥ Ï≤ôÎèÑÎ•º ÌÜµÌï¥ ÌäπÏ†ï ÏÉâÏÉÅÏù¥ Ïñ¥Îñ§ Í∞íÏùò Î≤îÏúÑÎ•º ÎÇòÌÉÄÎÇ¥ÎäîÏßÄ Ïïå Ïàò ÏûàÏäµÎãàÎã§.
#
# Í∞í (Value):
# Í∞Å ÏÖÄÏóê ÌëúÏãúÎê† Ïàò ÏûàÎäî Ïã§Ï†ú Îç∞Ïù¥ÌÑ∞ Í∞íÏûÖÎãàÎã§. (ÏÑ†ÌÉù ÏÇ¨Ìï≠)
#
# ## ÌûàÌä∏ÎßµÏùò Ï£ºÏöî Ïö©ÎèÑ
# ÌûàÌä∏ÎßµÏùÄ Î≥µÏû°Ìïú Îç∞Ïù¥ÌÑ∞ ÏÜçÏóêÏÑú
# Ìå®ÌÑ¥ÏùÑ ÏßÅÍ¥ÄÏ†ÅÏúºÎ°ú Î∞úÍ≤¨ÌïòÎäî Îç∞ Îß§Ïö∞ Ïú†Ïö©Ìï©ÎãàÎã§.
#
# ÏÉÅÍ¥ÄÍ¥ÄÍ≥Ñ Î∂ÑÏÑù
# Ïó¨Îü¨ Î≥ÄÏàòÎì§ ÏÇ¨Ïù¥Ïùò ÏÉÅÍ¥ÄÍ¥ÄÍ≥ÑÎ•º ÌñâÎ†¨Î°ú ÎßåÎì† ÌõÑ ÌûàÌä∏ÎßµÏúºÎ°ú ÏãúÍ∞ÅÌôîÌïòÎ©¥,
# Ïñ¥Îñ§ Î≥ÄÏàòÎì§ÎÅºÎ¶¨ Í∞ïÌïú Ïñë ÎòêÎäî ÏùåÏùò ÏÉÅÍ¥ÄÍ¥ÄÍ≥ÑÎ•º Í∞ñÎäîÏßÄ ÏâΩÍ≤å ÌååÏïÖÌï† Ïàò ÏûàÏäµÎãàÎã§.
#
# ÏõπÏÇ¨Ïù¥Ìä∏ ÏÇ¨Ïö©Ïûê ÌñâÎèô Î∂ÑÏÑù
# ÏõπÌéòÏù¥ÏßÄÏóêÏÑú ÏÇ¨Ïö©ÏûêÎì§Ïù¥ ÎßàÏö∞Ïä§Î•º ÎßéÏù¥ ÏõÄÏßÅÏù¥Í±∞ÎÇò
# ÌÅ¥Î¶≠ÌïòÎäî ÏòÅÏó≠ÏùÑ Î∂âÏùÄÏÉâÏúºÎ°ú ÌëúÏãúÌïòÏó¨,
# Ïñ¥Îñ§ Î∂ÄÎ∂ÑÏóê Í¥ÄÏã¨Ïù¥ ÏßëÏ§ëÎêòÎäîÏßÄ Î∂ÑÏÑùÌï† Ïàò ÏûàÏäµÎãàÎã§.
#
# ÏßÄÎ¶¨Ï†Å Îç∞Ïù¥ÌÑ∞ ÏãúÍ∞ÅÌôî
# ÏßÄÎèÑ ÏúÑÏóê ÏßÄÏó≠Î≥Ñ Ïù∏Íµ¨ Î∞ÄÎèÑÎÇò
# ÏßàÎ≥ë Î∞úÏÉùÎ•† Í∞ôÏùÄ Îç∞Ïù¥ÌÑ∞Î•º ÏÉâÏÉÅÏúºÎ°ú ÌëúÌòÑÌïòÏó¨ ÏßÄÏó≠Ï†Å Ìé∏Ï∞®Î•º ÏâΩÍ≤å ÌôïÏù∏Ìï† Ïàò ÏûàÏäµÎãàÎã§.

# 6. ÌûàÌä∏Îßµ
import seaborn as sns
data = np.random.rand(5, 5)
plt.figure(figsize=(8, 6))
sns.heatmap(data, annot=True, cmap="coolwarm")
plt.title("ÌûàÌä∏Îßµ ÏòàÏ†ú", fontproperties=font_prop)
plt.show()


# Ï£ºÏöî ÏòµÏÖò
#
# annot=True ‚Üí Í∞Å ÏÖÄÏóê Í∞í ÌëúÏãú
# cmap="coolwarm" ‚Üí ÏÉâÏÉÅ ÌÖåÎßà Î≥ÄÍ≤Ω
#
# üìå Ï∂îÍ∞Ä ÏòµÏÖò
#
# linewidths=0.5 ‚Üí ÏÖÄ Í∞Ñ Í≤ΩÍ≥ÑÏÑ† ÏÑ§Ï†ï
# linecolor='black' ‚Üí Í≤ΩÍ≥ÑÏÑ† ÏÉâÏÉÅ ÏßÄÏ†ï
# cbar_kws={'shrink': 0.8} ‚Üí Ïª¨Îü¨Î∞î ÌÅ¨Í∏∞ Ï°∞Ï†à
# annot_kws={'size': 12} ‚Üí ÌÖçÏä§Ìä∏ ÌÅ¨Í∏∞ Ï°∞Ï†à
# xticklabels=['A', 'B', 'C', 'D', 'E'] ‚Üí XÏ∂ï Î†àÏù¥Î∏î Î≥ÄÍ≤Ω
# yticklabels=['V1', 'V2', 'V3', 'V4', 'V5'] ‚Üí YÏ∂ï Î†àÏù¥Î∏î Î≥ÄÍ≤Ω
# mask=np.triu(np.ones_like(data, dtype=bool)) ‚Üí ÏÉÅÏÇºÍ∞ÅÌòï ÎßàÏä§ÌÇπ
# center=0 ‚Üí Ï§ëÏã¨Í∞í ÏÑ§Ï†ï (0ÏùÑ Í∏∞Ï§ÄÏúºÎ°ú ÏÉâÏÉÅ ÌëúÌòÑ)
# cmap='coolwarm' ‚Üí ÏÉâÏÉÅ ÌÖåÎßà Î≥ÄÍ≤Ω
# plt.xticks(rotation=45) ‚Üí XÏ∂ï Î†àÏù¥Î∏î ÌöåÏ†Ñ
```

```python
# ÏòàÏãú 9
# ‚ë° ÏãúÍ≥ÑÏó¥ Îç∞Ïù¥ÌÑ∞ Î∂ÑÏÑù (Ïù¥Îèô ÌèâÍ∑†ÏÑ† Ï∂îÍ∞Ä)
# Ïù¥ÎèôÌèâÍ∑†ÏÑ†(Moving Average, MA)ÏùÄ
# Ï£ºÏãù Ï∞®Ìä∏ Îì±ÏóêÏÑú Îã®Í∏∞Ï†ÅÏù∏ Í∞ÄÍ≤© Î≥ÄÎèô(ÎÖ∏Ïù¥Ï¶à)ÏùÑ Ï†úÍ±∞ÌïòÍ≥†
# Ï†ÑÎ∞òÏ†ÅÏù∏ Ï∂îÏÑ∏Ïùò Î∞©Ìñ•ÏùÑ ÌååÏïÖÌïòÍ∏∞ ÏúÑÌï¥ ÏÇ¨Ïö©ÎêòÎäî Í∞ÄÏû• Í∏∞Î≥∏Ï†ÅÏù∏ Í∏∞Ïà†Ï†Å ÏßÄÌëúÏûÖÎãàÎã§. üìà

# Ïù¥Î¶Ñ Í∑∏ÎåÄÎ°ú,
# ÏùºÏ†ï Í∏∞Í∞Ñ ÎèôÏïàÏùò Í∞ÄÍ≤©ÏùÑ ÏÇ∞Ïà† ÌèâÍ∑†Ìïú Í∞íÏùÑ ÏÑ†ÏúºÎ°ú Ïó∞Í≤∞Ìïú Í≤ÉÏûÖÎãàÎã§.
# "Ïù¥Îèô(Moving)"Ïù¥ÎùºÎäî Îã®Ïñ¥Í∞Ä Î∂ôÎäî Ïù¥Ïú†Îäî
# ÏÉàÎ°úÏö¥ ÎÇ†Ïùò Îç∞Ïù¥ÌÑ∞Í∞Ä Ï∂îÍ∞ÄÎêòÎ©¥ Í∞ÄÏû• Ïò§ÎûòÎêú ÎÇ†Ïùò Îç∞Ïù¥ÌÑ∞Îäî
# Ï†úÏô∏ÎêòÎäî Î∞©ÏãùÏúºÎ°ú ÌèâÍ∑†Ïù¥ Í≥ÑÏÜçÌï¥ÏÑú Ïù¥ÎèôÌïòÍ∏∞ ÎïåÎ¨∏ÏûÖÎãàÎã§.

# ÌïµÏã¨ ÏõêÎ¶¨
# Ïù¥ÎèôÌèâÍ∑†ÏÑ†ÏùÄ Í≥ÑÏÇ∞ Î∞©ÏãùÏù¥ Îß§Ïö∞ Í∞ÑÎã®Ìï©ÎãàÎã§.
#
# 5Ïùº Ïù¥ÎèôÌèâÍ∑†ÏÑ†: ÏµúÍ∑º 5ÏùºÍ∞ÑÏùò Ï¢ÖÍ∞ÄÎ•º Î™®Îëê ÎçîÌïú ÌõÑ 5Î°ú ÎÇòÎàà Í∞í.
#
# 20Ïùº Ïù¥ÎèôÌèâÍ∑†ÏÑ†: ÏµúÍ∑º 20ÏùºÍ∞ÑÏùò Ï¢ÖÍ∞ÄÎ•º Î™®Îëê ÎçîÌïú ÌõÑ 20ÏúºÎ°ú ÎÇòÎàà Í∞í.
#
# Ïù¥ Í≥ºÏ†ïÏùÑ Îß§Ïùº Î∞òÎ≥µÌïòÏó¨ Í≥ÑÏÇ∞Îêú ÌèâÍ∑†Í∞íÎì§ÏùÑ
# ÏÑ†ÏúºÎ°ú Ïù¥ÏúºÎ©¥ Ïù¥ÎèôÌèâÍ∑†ÏÑ†Ïù¥ ÎßåÎì§Ïñ¥ÏßëÎãàÎã§.
# Ïù¥ ÏÑ†ÏùÄ Îß§ÏùºÏùò Ï£ºÍ∞Ä Îì±ÎùΩÎ≥¥Îã§ Ìõ®Ïî¨ Î∂ÄÎìúÎüΩÍ≤å ÏõÄÏßÅÏó¨, ÏãúÏû•Ïùò ÌÅ∞ ÌùêÎ¶ÑÏùÑ Î≥¥Ïó¨Ï§çÎãàÎã§.

# 8. ÏãúÍ≥ÑÏó¥ Îç∞Ïù¥ÌÑ∞ Î∂ÑÏÑù
import numpy as np
import pandas as pd

data = np.cumsum(np.random.randn(100)) # 100Í∞úÏùò ÎûúÎçòÌï® Ïã§Ïàò
window = 10 # Ïù¥Îèô ÌèâÍ∑†ÏÑ†, 10ÏùºÏùò Ïù¥Îèô ÌèâÍ∑†ÏÑ†
rolling_mean = pd.Series(data).rolling(window=window).mean()
# Îç∞Ïù¥ÌÑ∞Ïóê **Ïù¥ÎèôÏ∞Ω(rolling window)**ÏùÑ Ï†ÅÏö©Ìï©ÎãàÎã§.
# window=10ÏúºÎ°ú ÏÑ§Ï†ïÌñàÏúºÎØÄÎ°ú,
# Îç∞Ïù¥ÌÑ∞Î•º 10Í∞úÏî© Î¨∂Ïñ¥ ÌïòÎÇòÏùò Ï∞ΩÏúºÎ°ú ÎßåÎì≠ÎãàÎã§.
# Ï≤´ Î≤àÏß∏ Ï∞ΩÏùÄ 0~9Î≤à Îç∞Ïù¥ÌÑ∞,
# Îëê Î≤àÏß∏ Ï∞ΩÏùÄ 1~10Î≤à Îç∞Ïù¥ÌÑ∞,
# ... ÏôÄ Í∞ôÏù¥ Ìïú Ïπ∏Ïî© Ïù¥ÎèôÌïòÎ©∞ Ï∞ΩÏùÑ Íµ¨ÏÑ±Ìï©ÎãàÎã§.

plt.figure(figsize=(10, 5))
plt.plot(data, label="ÏõêÎ≥∏ Îç∞Ïù¥ÌÑ∞", alpha=0.5)
plt.plot(rolling_mean, label=f"{window}Ïùº Ïù¥Îèô ÌèâÍ∑†", color="red")
plt.legend(prop=font_prop)
plt.title("ÏãúÍ≥ÑÏó¥ Îç∞Ïù¥ÌÑ∞ Î∂ÑÏÑù", fontproperties=font_prop)
plt.show()
```
